```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 6, 8)
res = np.array([1,3])

y1 = 3*x 
y2 = 1/2*x+6
y4 = x+2

plt.figure()
plt.plot(x, y1, label="y<=3x")
plt.plot(x, y2, label="y<=1/2x+6")
plt.plot(x, y4, label="y>=x+2")
plt.axvline(x=6, ymin=0, ymax=15, label="x<=6", c='pink')
plt.plot(res[0], res[1], "o")

z = np.minimum(y1, y2)
plt.fill_between(x, z, y4, where=z>y4, facecolor='red', alpha=0.3)
plt.ylim(0, 15)
plt.legend(loc=0)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
